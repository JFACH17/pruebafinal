```python
categorias = []
labels = []
imagenes = []
```

```python
categorias = os.listdir ('C:\\Users\\HP\\Documents\\Pycham\\pruebafinal\\CarneDataset\\train\\')
print (categorias)
```

```python
x=0
for directorio in categorias :
    for imagen in os.listdir ('C:\\Users\\HP\\Documents\\Pycham\\pruebafinal\\CarneDataset\\train\\'+ directorio):
        img = Image.open('C:\\Users\\HP\\Documents\\Pycham\\pruebafinal\\CarneDataset\\train\\'+ directorio +'\\'+imagen).resize((200,200))
        img = np.asarray(img)
        imagenes.append(img)
        labels.append(x)
    x +=1
    
print (labels)
```

```python
model= tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(200,200)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
labels = np.asarray(labels)
print(labels)
```

```python
model.fit(imagenes,labels,epochs=30)
```
